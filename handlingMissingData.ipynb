# Filling in missing values

```python
# many away to fill the missing value 
# fill all the missing value by '0'
#  fill by mean median mode
# fill by forward or backward
# fill by specific value
```

```python
# Step - 1: fill all missing values by 0

n_df = df.fillna(0)
n_df
```

```python
n_df.isnull().sum()
```

```python
# Fill numerical data with the mean
df['yacWPA'].fillna(df['yacWPA'].mean(), inplace=True)
df



```

```python
# Fill categorical data with the mode
df['airWPA'].fillna(df['airWPA'].mode()[0], inplace=True)
df
```

```python
# Fill numerical data with the median
df['yacEPA'].fillna(df['yacEPA'].median(), inplace=True)
df
```

```python
# Fill missing values with the previous row's value
df['down'].fillna(method='ffill', inplace=True)

df
```

```python
# Fill missing values with the next row's value
df['down'].fillna(method='bfill', inplace=True)
df
```
